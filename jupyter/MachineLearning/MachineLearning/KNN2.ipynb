If we have an unlabeled fruit with a known weight and a color category, then applying the k-nearest neighbor method (with any distance formula) will most likely find the nearest k neighbors (if they are green, red, or yellow, the unlabeled fruit is
most likely a pear, apple, or banana respectively). The following code demonstrates k-nearest neighbor algorithm using the shape and weight of fruits:

```python
import csv
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
```

```python
count=0
x=[]
y=[]
z=[]
```

```python
with open('fruits_data.csv','r') as csvf:
    reader = csv.reader(csvf, delimiter=',')
    for row in reader:
        if count>0:
            x.append(row[0])
            y.append(row[2])
            if(row[1] =='orange'): 
                z.append('r')
            elif(row[1] == 'lemon'): 
                z.append('g')
            else:
                z.append('b')
        count += 1
plt.figure(figsize=(11,11))

recs=[]
classes = ['Apples','Pear','Bananas']
class_colours = ['r','g','y']
plt.title("Apples, Bananas and Pear by Weight and Shape", fontsize=18)
plt.xlabel("Shape category number", fontsize=14)
plt.ylabel("Weight in ounces", fontsize=14)
plt.scatter(x,y, s=600, c=z)
plt.show()

```

```python
from math import pow, sqrt
from math import pow, sqrt
dist=[]
def determineFruit(xv, yv, threshold_radius):
    for i in range(1,len(x)):
        xdif=pow(float(x[i])-xv, 2)
        ydif=pow(float(y[i])-yv, 2)
        sqrtdist = sqrt(xdif+ydif)
        if ( xdif < threshold_radius and
            ydif < thresholdradius and sqrtdist < threshold_radius):dist.append(sqrtdist)
        else:
            dist.append(99)
    pear_count=0
    apple_count=0
    banana_count=0
    for i in range(1,len(dist)):
        if dist[i] < threshold_radius:
            if z[i] == 'g': pear_count += 1
            if z[i] == 'r': apple_count += 1
            if z[i] == 'y': banana_count += 1
    if ( apple_count >= pear_count and apple_count >= banana_count ):
        return "apple"
    elif ( pear_count >= apple_count and pear_count >= banana_count):
        return "lemon"
    elif ( banana_count >= apple_count and banana_count >= pear_count):
        return "orange"
dist=[]
determine = determineFruit(3.5,6.2, 1)
print(determine)
```
