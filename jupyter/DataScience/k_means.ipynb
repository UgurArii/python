```python
import numpy as np
import matplotlib.pyplot as plt
```

For every cluster, the average of the silhouette coefficient of all the points can serve as a
good measure of the cluster quality. An average of the silhouette coefficient of all the data
points can serve as an overall quality metric for the clusters formed.
Letâ€™s go ahead and generate some random data:

```python
def get_random_data():
    x_1 = np.random.normal(loc=0.2, scale=0.2, size=(100,100))
    x_2 = np.random.normal(loc=0.9, scale=0.1, size=(100,100))
    x = np.r_[x_1, x_2]
    return x
```

```python
x = get_random_data()
plt.cla()
plt.figure(1)
plt.title("Generated Data")
plt.scatter(x[:,0], x[:,1])
plt.show()
```
