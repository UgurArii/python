# Step 3 : Write the Basch Commands (80 Points)

Write the needed commands to accomplesh the task given above.

```python
#!/bin/bash

# Download the dataset
!wget https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data

# 1. Print the first five lines of the file
!echo "1. The first five lines of the file:"
!head -n 5 iris.data

# 2. Count the number of lines in the file
!echo "2. The number of lines in the file:"
!wc -l iris.data

# 3. Extract the second column of the file
!echo "3. The second column of the file:"
!cut -d ',' -f 2 iris.data

# 4. Calculate the average of each column in the file
!echo "4. The average of each column in the file:"
!awk -F ',' '{sum1+=$1; sum2+=$2; sum3+=$3; sum4+=$4; n+=1} END {print sum1/n "," sum2/n "," sum3/n "," sum4/n}' iris.data

# 5. Determine the species with the highest and lowest sepal lengths
!echo "5. The species with the highest and lowest sepal lengths:"
!awk -F ',' 'BEGIN {max=0; min=1000} {if ($1 > max) {max=$1; max_species=$5}; if ($1 < min) {min=$1; min_species=$5}} END {print "Max sepal length: " max_species " (" max ")" "\nMin sepal length: " min_species " (" min ")"}' iris.data

# 6. Determine the number of samples for each species
!echo "6. The number of samples for each species:"
!awk -F ',' '{species[$5]++} END {for (i in species) {print i, species[i]}}' iris.data

# 7. Generate a bar chart of the number of samples for each species using gnuplot
!echo "7. Generating a bar chart of the number of samples for each species using gnuplot..."
!awk -F ',' '{species[$5]++} END {for (i in species) {print i, species[i]}}' iris.data | gnuplot -e "set style data histogram; set style histogram cluster; set xtics rotate by -45; plot '-' using 2:xtic(1) title 'Number of samples per species'"

# 8. Backup the Iris dataset and the script itself
!echo "8. Backing up the Iris dataset and the script itself..."
!mkdir backup
!cp iris.data backup/
!cp iris.sh backup/

!echo "Data science operations completed successfully."
```
