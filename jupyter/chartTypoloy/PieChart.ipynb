```python
import matplotlib.pyplot as plt
labels = ['Nokia','Samsung','Apple','Lumia']
values = [10,30,45,15]
colors=['yellow','green','blue','red']
plt.pie(values, labels=labels, colors=colors)
plt.axis('equal')
```

```python
labels = ['Nokia','Samsung','Apple','Lumia']
values = [10,30,45,15]
colors=['yellow','green','blue','red']
explode = [0.3,0.4,0,0]
plt.pie(values, labels=labels, colors=colors, explode=explode, startangle=170)
plt.axis('equal')
```

```python
labels = ['Nokia','Samsung','Apple','Lumia']
values = [10,30,45,15]
colors=['yellow','green','blue','red']
explode = [0.3,0.4,0,0]
plt.title('A Pie Chart')
plt.pie(values, labels=labels, colors=colors, explode=explode, shadow=True, autopct='%1.2f%%',startangle=180)
plt.axis('equal')
```

```python
import pandas as pd
data = {'series1':[1,3,4,3,5],
       'series2':[2,4,5,2,4],
       'series3':[3,2,3,1,3]}
df = pd.DataFrame(data)
df['series1'].plot(kind='pie', figsize=(6,6))
```
