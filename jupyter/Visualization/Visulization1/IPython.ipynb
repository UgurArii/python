```python
import matplotlib.pyplot as plt
from numpy.random import rand
```

```python
fig = plt.figure()
fig.suptitle('IPython Plotting')
for color in ['orange','blue','red']:
    n = 400
    x, y = rand(2,n)
    scale = 200.0 * rand(n)
    plt.scatter(x, y, c=color, s=scale, label=color, alpha=0.5, edgecolors='none')
plt.legend()
plt.grid(True)
plt.show()
```
